# ETS Decomposition
**Consiste en descomponer una serie temporal en:**
* Componente de **Error o Residual**
* Componente **Tendencial**
* Componente **Estacionaria o Seasonal**

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('../../Resources/Data/airline_passengers.csv', index_col='Month', parse_dates=True)
data.dropna(inplace=True)
data.head()
```

```python
data.plot(figsize=(12,5))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(data['Thousands of Passengers'], model='multiplicative')
result.plot(); # Si no se pone el ; se plotea 2 veces :/
```

**Si queremos que se plotee con un figsize distinto...**

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12, 5

result.plot();
```

**Para plotear sólamente la componente tendencial...**

```python
result.trend.plot(figsize=(12, 6))
```
