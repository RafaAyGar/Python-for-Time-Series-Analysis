# Vamos a descomponer una serie temporal en su componente cíclica y tendencial usando el HP Filter
## Luego visualizaremos la componente tendencial junto con los datos iniciales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../../Resources/Data/macrodata.csv', index_col=0, parse_dates=True)
data['realgdp'].plot()
```

### **Vamos a usar el filtro de Hodrick-Prescott (HP Filter) para separar la componente cíclica de la tendencial**

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

**El smoothing parameter lo ponemos a 1600 porque es el valor estándar para datos trimestrales**

```python
gdp_cycle, gdp_trend = hpfilter(data['realgdp'], lamb=1600)
```

**Vamos a representar los datos iniciales de realgdp, con la componente de tendencia**

```python
data['gdptrend'] = gdp_trend
data[['realgdp', 'gdptrend']].plot(figsize=(12,5))
plt.show()
```

**Vamos a ampliar para ver de forma más clara lo que pasó en los últimos 5 años del dataset**

```python
data[['gdptrend', 'realgdp']]['2004-01-01':].plot(figsize=(12, 5))
```
