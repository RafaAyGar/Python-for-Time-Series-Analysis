# Moving Average Model

**En este cuaderno vamos a ver el modelo de medias móviles, para que su utilidad quede plasmada de forma más cara, vamos a generar nuestro propio dataset y más tarde veremos cómo montar y evaluar el modelo**
**Antes que nada, la estructura del modelo, por ejemplo de orden 1 es:**
## $Y_t = \mu + \phi\cdot\epsilon_{t-1}$
**Donde $\epsilon \sim N(0, \sigma)$**

**El término de error $\epsilon$ será sintético, es decir, generado por nosotros**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Generamos el término de error $\epsilon$
**En este caso haremos que la varianza de la distribución normal que sigue $\epsilon$ sea 1, es decir, $\epsilon \sim N(0, 1)$**

```python
error = np.random.normal(0, 1, 300)
```

## Generamos el dataset a partir de $\epsilon$, estableciendo la media $\mu$ a 50

**La serie generada seguirá la siguiente función:** 
### $y_t = 50 + 0.4\epsilon_{t-1} + 0.3\epsilon_{t-2} + \epsilon_{t}$

```python
date_index = pd.date_range(start="2019-9-1", end="2020-1-1")
mean = 50
ts = [] # Time Series
for t in range(1, len(date_index)+1):
    value = mean + 0.4*error[t-1] + 0.2*error[t-2] + error[t]
    ts.append(value)
ts = pd.Series(ts, date_index)
ts = ts.asfreq(pd.infer_freq(ts.index))
```

## La serie "ts" es ahora nuestra serie temporal generada a partir de la señal de error $\epsilon$, vamos a plotearla para visualizarla mejor

```python
plt.figure(figsize=(10, 6))
ts.plot()
plt.axhline(mean, linestyle="--", color="grey")
```

## Ahora vamos a mostrar los ACF y PACF (Autocorrelation y Partial Autocorrelation Functions)

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
acf_vals = acf(ts)
num_lags = 10
plt.bar(range(num_lags), acf_vals[:num_lags])
```

```python
pacf_vals = pacf(ts)
num_lags = 10
plt.bar(range(num_lags), pacf_vals[:num_lags])
```

## Los ploteos revelan que la correlación con los periodos anteriores, es muy fuerte en los más cercanos, pero se reduce de forma consistente en el resto, es nos dice que la serie temporal puede modelarse bastante bien con un MA Model

## Lo siguiente es separar los datos en train y test, y crear y poner a prueba el modelo

```python
from datetime import datetime, timedelta
```

```python
train_end_date = datetime(2019,12,28)
train_data = ts[:train_end_date]
test_data = ts[train_end_date + timedelta(days=1):]
test_data
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

## ¡Ojo! Con order(0,0,2), estamos indicando con el primer 0 que es un modelo no autorregresivo, y con el 2 final, que vamos a utilizar los términos de error de los 2 últimos periodos $\epsilon_{t-1}$, $\epsilon_{t-2}$

```python
model = ARIMA(train_data, order=(0,0,2))
model = model.fit()
print(model.summary())
```

## Observamos que el modelo entrenado se ajusta bastante a la función que inicialmente escogimos para generar la serie temporal, vemos que el término independiente es muy muy cercano a 50, así como los coeficientes son cercanos a 0.4, y a 0.2 respectivamente
## Modelo entrenado:

### $y_t = 50.36 + 0.4301\epsilon_{t-1} + 0.2866\epsilon_{t-2}$
## A continuación se harán las predicciones correspondientes a los datos de testeo, para ver cómo rinde el modelo

```python
predictions_start_date = test_data.index[0]
predictions_end_date = test_data.index[-1]
```

```python
predictions = model.predict(start=predictions_start_date, end=predictions_end_date)
predictions.plot()
test_data.plot()
```

## Si observamos la escala del eje y del plot, descubrimos que nuestro modelo hace unas predicciones relativamente ajustadas
## Para verlo con más claridad, vamos a calcular el porcentaje medio de error

```python
residuals = test_data - predictions
absolute_percent_error = round(np.mean(abs(residuals/test_data)), 2)
print(f"El porcentaje de error medio cometido por el modelo es de {absolute_percent_error}%")
```
